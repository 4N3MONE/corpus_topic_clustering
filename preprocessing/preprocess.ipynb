# NP

### 문제: xml파일의 스페셜 캐릭터를 본문에 적어둬서 파싱을 실패한다.  
### 해결: raw string으로 읽어 manual하게 처리하기

```python
rootpath = "data/NP"
filelist = os.listdir(rootpath)
```

```python
files = []

for file in filelist:
    files.append(os.path.join(rootpath, file))
```

```python
def extract_content_from_(file_name):
    with open(file_name, 'r') as f:
        ws = f.read()
        start = -1
        for i, char in enumerate(ws[40:]):
            if(ws[i-3:i] == "<p>"):
                start = i
                break
        result = ws[start:-20] 
    result = re.sub('</?[a-z]?>', '', result)
    result = re.sub('\s+', ' ', result)
    result = re.sub('&[a-z]+;', '', result)
    result = re.sub('&[가-힣0-9a-zA-Z]+&', '', result)
    return result
```

```python
extract_content_from_(files[0])
```

```python
results = []

for file in files:
    result = extract_content_from_(file)
    results.append(result)

len(results)
```

```python
np_data = {'data' : []}
corpus_name = 'np'
for idx, result in enumerate(results):
    np_data['data'].append({
        '_id' : '_'.join([corpus_name, str(idx)]),
        'text' : result
    })
```

```python
np_data
```

```python
import json

with open('result/np_data.json', 'w') as f:
    json.dump(np_data, f, ensure_ascii=False)
    
```

# NEWS

```python
news_root = 'data/origin/News'
news_files = [os.path.join(news_root, filename) for filename in os.listdir(news_root)]
news_result = {'data' : []}
news_task_name = 'news'
idx = 0
for news_file in news_files:
    with open(news_file, 'r') as f:
        news_file = json.load(f)
    for document in news_file['document']:
        article = []
        for paragraph in document['paragraph']:
            article.append(paragraph['form'])
        news_result['data'].append({
                '_id' : '_'.join([news_task_name, str(idx)]),
                'text' : ' '.join(article)
        })
        idx += 1

print(len(news_result['data']))
news_result['data'][:3]
```
